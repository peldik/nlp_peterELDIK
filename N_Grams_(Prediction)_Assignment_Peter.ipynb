```python
import nltk
from nltk.corpus import brown
from nltk.util import ngrams
from collections import Counter, defaultdict
import string
import pandas as pd


nltk.download('brown')
nltk.download('punkt')


tokenized_text = [word.lower() for word in brown.words() if word.isalpha()]


def get_ngram_df(tokens, n=2):
    n_grams = list(ngrams(tokens, n))
    freq = Counter(n_grams)
    df = pd.DataFrame(freq.items(), columns=[f'{n}-gram', 'Frequency']).sort_values(by='Frequency', ascending=False)
    return df


bigram_df = get_ngram_df(tokenized_text, n=2)
print("Top 5 bigrams:")
display(bigram_df.head())

def next_word_predictions(tokens, n, context, top_k=5):
    """
    Predict next words based on n-gram model.
    - tokens: list of words
    - n: n-gram size (e.g. 2 for bigram, 3 for trigram)
    - context: input word(s), as a string
    - top_k: number of suggestions to return
    """
    context_tokens = nltk.word_tokenize(context.lower())
    context_tokens = [w for w in context_tokens if w.isalpha()]
    if len(context_tokens) != n - 1:
        print(f"[!] Please provide {n-1} word(s) for an n={n} model.")
        return []

    n_grams = list(ngrams(tokens, n))
    predictions = defaultdict(int)

    for gram in n_grams:
        if list(gram[:-1]) == context_tokens:
            predictions[gram[-1]] += 1

    sorted_preds = sorted(predictions.items(), key=lambda x: x[1], reverse=True)
    return sorted_preds[:top_k]

print("\nTop next words for 'the':")
predictions = next_word_predictions(tokenized_text, n=2, context="the", top_k=5)
for word, freq in predictions:
    print(f"- {word} ({freq} times)")

print("\nTop next words for 'in the':")
predictions = next_word_predictions(tokenized_text, n=3, context="in the", top_k=5)
for word, freq in predictions:
    print(f"- {word} ({freq} times)")
```
