```python
import pandas as pd
import re
from transformers import pipeline
```

```python
tweetsdf = pd.read_csv("tweets-data (1).csv")
```

```python
sampledf = tweetsdf.sample(n=500, random_state=42).copy()
```

```python

def clean(text):
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'[^A-Za-z0-9\s]', '', text)
    text = text.lower().strip()
    return text
```

```python
sampledf['cleaned'] = sampledf['Tweets'].astype(str).apply(clean)
```

```python
sentimentpipeline = pipeline("sentiment-analysis", truncation=True)
```

```python
results = sentimentpipeline(sampledf['cleaned'].tolist(), batch_size=32)

sampledf['sentimentLabel'] = [res['label'] for res in results]
sampledf['sentimentScore'] = [res['score'] for res in results]


sampledf[['Tweets', 'cleaned', 'sentimentLabel', 'sentimentScore']].head()
```
